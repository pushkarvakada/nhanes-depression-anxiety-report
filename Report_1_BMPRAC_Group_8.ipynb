## 1.1 What is NHANES?
The National Health and Nutrition Examination Survey (NHANES) is a program conducted by the U.S. CDC which collects data from a representative sample of the American population. ## 1.2 Our Research Question & Definitions
**Research Question:**  
How do gender, sleep duration, and demographic factors influence three key health indicators ‚Äî PHQ-9 depression scores, resting heart rate, and Creatine levels ‚Äî in U.S. adults aged 20 and above?

Definitions;

**What is PHQ - 9?**

The Patient Health Questionnaire-9 (PHQ-9) is a validated tool for measuring the severity of depression symptoms in population-based studies. NHANES (National Health and Nutrition Examination Survey) provides access to this measure along with a wide range of related behavioral, physiological, and demographic data in a representative U.S. sample.

We included **resting heart rate (BPXPLS)** and **creatinine levels (URXCRS)** as additional biomarkers to investigate potential biological associations with mental health.

- **Heart rate** is a key physiological indicator of autonomic nervous system activity. Research has shown that elevated resting heart rate is associated with stress, anxiety, and poor emotional regulation. In some cases, higher heart rate can reflect heightened sympathetic arousal in individuals experiencing psychological distress.

- **Creatinine** is a waste product produced by muscles and excreted by the kidneys. Though not a direct marker of mental health, abnormal creatinine levels may signal underlying health conditions or chronic stress, which can affect mood, energy levels, and overall well-being. Some studies also link depression with altered kidney function and muscle mass, making creatinine a potential indirect indicator.

**Importing Libraries, Mounting Files from Drive & Providing Path**

```python
from google.colab import drive
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")

drive.mount('/content/drive')

# --- Define Paths ---
drive_path = '/content/drive/MyDrive/NHANES_Data/'
```

**Loading & Merging the NHANES Cycles (DEMO, DPQ, SLQ & BPX)**

```python
cycles = {
    '07-08': 'E',
    '09-10': 'F',
    '11-12': 'G',
    '13-14': 'H',
    '15-16': 'I',
    '17-18': 'J'
}

# --- Load datasets ---
demo, dpq, slq, bpx, biopro = {}, {}, {}, {}, {}

for c, s in cycles.items():
    try:
        demo[c]    = pd.read_sas(drive_path + f'DEMO_{s}.xpt')
        dpq[c]     = pd.read_sas(drive_path + f'DPQ_{s}.xpt')
        slq[c]     = pd.read_sas(drive_path + f'SLQ_{s}.xpt')
        bpx[c]     = pd.read_sas(drive_path + f'BPX_{s}.xpt')
        biopro[c]  = pd.read_sas(drive_path + f'BIOPRO_{s}.xpt')
    except Exception as e:
        print(f"Error loading {c}: {e}")
```

**Merge ALL Cycles into a fully merged dataset df**

```python
merged = []
for c in cycles:
    if all(c in d for d in [demo, dpq, slq, bpx, biopro]):
        temp = demo[c].merge(dpq[c], on='SEQN')\
                      .merge(slq[c], on='SEQN')\
                      .merge(bpx[c], on='SEQN')\
                      .merge(biopro[c], on='SEQN')
        temp['Cycle'] = c
        merged.append(temp)

df = pd.concat(merged, ignore_index=True)
```

**Define cleaned data df_clean for analysis**


```python
# --- PHQ-9 item codes ---
phq_cols = ['DPQ010','DPQ020','DPQ030','DPQ040','DPQ050',
            'DPQ060','DPQ070','DPQ080','DPQ090']

# --- Step 1: Remove invalid PHQ-9 responses (7, 9) ---
df = df[~df[phq_cols].isin([7, 9]).any(axis=1)]

# --- Step 2: Drop invalid education codes (7, 9) ---
df = df[~df['DMDEDUC2'].isin([7, 9])]

# --- Step 3: Drop invalid race codes (keep only 1‚Äì5) ---
df = df[df['RIDRETH1'].isin([1, 2, 3, 4, 5])]

# --- Step 4: Drop invalid gender codes (keep 1 or 2) ---
df = df[df['RIAGENDR'].isin([1, 2])]

# --- Step 5: Drop invalid sleep hours (< 3 or > 14) ---
df = df[(df['SLD010H'] >= 3) & (df['SLD010H'] <= 14)]

# --- Step 6: Drop invalid heart rate (30‚Äì150 bpm) ---
df = df[(df['BPXPLS'] >= 30) & (df['BPXPLS'] <= 150)]

# --- Step 7: Drop invalid creatinine (0.3‚Äì2.0 mg/dL) ---
df = df[(df['LBXSCR'] >= 0.3) & (df['LBXSCR'] <= 2.0)]

# --- Step 8: Drop rows with missing values in any key variable ---
required_cols = phq_cols + [
    'SLD010H', 'RIAGENDR', 'RIDAGEYR', 'RIDRETH1',
    'DMDEDUC2', 'INDFMPIR', 'BPXPLS', 'LBXSCR'
]
df = df.dropna(subset=required_cols)

# --- Step 9: Filter adults (‚â• 20 years) ---
df = df[df['RIDAGEYR'] >= 20]

# --- Step 10: Calculate PHQ-9 and convert to int ---
df['PHQ9'] = df[phq_cols].sum(axis=1).astype(int)

# --- Step 11: Rename for clarity ---
df_clean = df.rename(columns={
    'SLD010H': 'SleepHours',
    'RIAGENDR': 'Gender',
    'RIDRETH1': 'Race',
    'INDFMPIR': 'IncomeRatio',
    'DMDEDUC2': 'EducationLevel',
    'BPXPLS': 'HeartRate',
    'LBXSCR': 'Creatinine'
})

df_clean['Gender'] = df_clean['Gender'].replace({1: 'Male', 2: 'Female'})

# --- Final check ---
print("‚úÖ FULLY CLEANED DATASET SHAPE:", df_clean.shape)
df_clean.head()
```
